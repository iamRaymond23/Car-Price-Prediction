```python
#data preparation
a=df['enginesize']
b=df['price']

print(stats.shapiro(a))
print(stats.shapiro(b))
# pval>0.05
# Data is normal

# Ho: corr =0 (no relation)
# ha: corr!= 0 (Relation)
t_stat,pval=stats.pearsonr(a,b)
t_stat,pval

sig_lvl=0.05
if pval<sig_lvl:
    print('Ha is selected - relation is there')
else:
    print('Ho is selected -  no relation')
```

2 sample TTEst

```python
df.columns
```

```python
# Data preparation

comp = df[df['enginelocation']=='front']['price']
no_comp =df[df['enginelocation']=='rear']['price']

# Finding_bar
x1_bar = np.mean(comp)
x2_bar = np.mean(no_comp)
print(x1_bar,x2_bar)
print(stats.shapiro(comp))
print(stats.shapiro(no_comp))
tsat,two_sid_prob = stats.ttest_ind(comp,no_comp)
print(tsat,two_sid_prob)

# inference
# Ho : mu1=mu2 (no relation)
# ha : mu1 != mu2 (relation)
sig_lvl = 0.05
if two_sid_prob<sig_lvl:
    print('Ha is selected- Related')
else:
    print('Ho is selected - Not related')
```

Two Sample TTest --- EngineLocation is related to price

```python
df['carbody'].value_counts()
```

```python
# Data Preparation

grp_A = df[df['carbody']=='sedan']['price']
grp_B = df[df['carbody']=='hatchback']['price']
grp_C = df[df['carbody']=='wagon']['price']
grp_D = df[df['carbody']=='hardtop']['price']
grp_E = df[df['carbody']=='convertible']['price']

# Test of normality

# Ho : Normal(skew=0)
# Ha : Not normal(skew!=0)

print(stats.shapiro(grp_A))
print(stats.shapiro(grp_B))
print(stats.shapiro(grp_C))
print(stats.shapiro(grp_D))
print(stats.shapiro(grp_E))
# pval>sig_lvl
# Data is normal

# Equality of Variance

# Ho : All variance are equal
# Ha : Atlest on variance is not equal
stats.levene(grp_A,grp_B,grp_C,grp_D,grp_E)

# pval>siglvl (siglvl=0.05)
# All variance are equal

# Assumption

# 1) Data is normal
# 2) Data has equal variance

# Hypothesis

# Ho : All pop means are equal
# Ha : Atleast one pop mean is not equal

# Test : ANOVA (more than two samples)

f_stat,pval = stats.f_oneway(grp_A,grp_B,grp_C,grp_D,grp_E)
print(f_stat,pval)

sig_lvl=0.05
if pval<sig_lvl:
    print('Ha is selected- related')
else:
    print('Ho is selected - not related')
```
